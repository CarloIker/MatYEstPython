#### y = x^2 + 3
### En el punto (1,4). Dibuja la parábola y muestra un segmento de la recta tangente en el punto (1,4).

##### PROCEDIMIENTO
y = x^2 + 3

- Derivamos la función de la parabola y sustituimos el valor "x" de la coordenada.
    * y' = 2x
    * y' = 2(1) = 2

```python
x = np.linspace(-6,6)
y_p = x**2 + 3
y_r = 2*x + 2
plt.plot(x, y_p, color="darkblue")
plt.plot(x, y_r, color="blue")
plt.plot(1,4,'o', color='gold')
plt.text(1, 4, "(1, 4)", ha='right', va='bottom')
plt.legend(("Parábola","Recta pendiente"))
plt.grid(True)
plt.title("EC. RECTA PENDIENTE DE PARÁBOLA")
plt.show()
```
